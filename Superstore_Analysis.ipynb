```python
# Profit by Category
category_profit = df.groupby("Category")["Profit"].sum().sort_values(ascending=False)
print("\nProfit by Category:\n", category_profit)
```

```python
# 6. Visualization: Profit by Category
category_profit.plot(kind="bar", color="#00994C", figsize=(6,4))
plt.title("Profit by Category",fontsize=14, weight="bold")
plt.ylabel("Profit ($)")
plt.xlabel("Category")
plt.xticks(rotation=0)
plt.grid(False) 

# Add data labels on top of bars
for i, value in enumerate(category_profit):
    plt.text(i, value, f"{value:,.0f}", ha="center", va="bottom", fontsize=10, weight="bold")

plt.show()
```
